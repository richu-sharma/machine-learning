```python
# Food Delivery Time Prediction
# Goal: Predict delivery time based on distance, traffic, and preparation time.
# Steps to Perform:
# • Create input features (distance, traffic, prep time)
# • Define a simple neural network
# • Use MSELoss for regression
# • Train the model using optimizer
# • Print loss after some epochs
# Sample Code:
# inputs = torch.randn(100, 3)
# targets = (inputs[:,0]*10 + inputs[:,1]*5 + inputs[:,2]*8).unsqueeze(1)
import torch
import torch.nn as nn
import torch.optim as optim

# Step 1: Create input features (distance, traffic, prep time)
inputs = torch.randn(100, 3)

# Target = weighted sum (simulating real delivery time)
targets = (inputs[:, 0]*10 + inputs[:, 1]*5 + inputs[:, 2]*8).unsqueeze(1)

# Step 2: Define a simple neural network
model = nn.Sequential(
    nn.Linear(3, 5),   # input layer → hidden layer
    nn.ReLU(),
    nn.Linear(5, 1)    # hidden → output
)

# Step 3: Define loss function (MSE for regression)
criterion = nn.MSELoss()

# Step 4: Define optimizer
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Step 5: Training loop
epochs = 100

for epoch in range(epochs):
    # Forward pass
    outputs = model(inputs)
    loss = criterion(outputs, targets)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print loss every 10 epochs
    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
# 2. Student Pass/Fail Predictor

# Goal: Predict whether student passes or fails.
# Steps to Perform:
# • Prepare input features
# • Convert target into 0/1
# • Use Sigmoid activation
# • Use BCELoss
# • Train model and observe loss
# Sample Code:
# inputs = torch.randn(100, 2)
# targets = ((inputs[:,0] + inputs[:,1]) > 0).float().unsqueeze(1)
import torch
import torch.nn as nn
import torch.optim as optim

# Step 1: Prepare input features
inputs = torch.randn(100, 2)

# Step 2: Convert target into 0/1 (Pass = 1, Fail = 0)
targets = ((inputs[:, 0] + inputs[:, 1]) > 0).float().unsqueeze(1)

# Step 3: Define model with Sigmoid activation
model = nn.Sequential(
    nn.Linear(2, 4),   # input → hidden
    nn.ReLU(),
    nn.Linear(4, 1),   # hidden → output
    nn.Sigmoid()       # for binary classification
)

# Step 4: Loss function (Binary Cross Entropy)
criterion = nn.BCELoss()

# Step 5: Optimizer
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Step 6: Training loop
epochs = 100

for epoch in range(epochs):
    # Forward pass
    outputs = model(inputs)
    loss = criterion(outputs, targets)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
# Weather Classification

# Goal: Predict weather category (0,1,2).
# Steps to Perform:
# • Prepare input features
# • Use integer targets (0,1,2)
# • Use CrossEntropyLoss
# • Train model
# • Check predictions
# Sample Code:
# inputs = torch.randn(100, 2)
# # targets = torch.randint(0,3,(100,))
import torch
import torch.nn as nn
import torch.optim as optim

# Step 1: Prepare input features
inputs = torch.randn(100, 2)

# Step 2: Integer targets (3 classes: 0,1,2)
targets = torch.randint(0, 3, (100,))

# Step 3: Define model (NO Sigmoid/Softmax here)
model = nn.Sequential(
    nn.Linear(2, 5),   # input → hidden
    nn.ReLU(),
    nn.Linear(5, 3)    # hidden → 3 output classes
)

# Step 4: Loss function (CrossEntropyLoss)
criterion = nn.CrossEntropyLoss()

# Step 5: Optimizer
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Step 6: Training loop
epochs = 100

for epoch in range(epochs):
    # Forward pass
    outputs = model(inputs)
    loss = criterion(outputs, targets)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# Step 7: Check predictions
with torch.no_grad():
    predictions = torch.argmax(model(inputs), dim=1)
    print("\nSample Predictions:", predictions[:10])
```

```python
# 4. Movie Rating Predictor
# Goal: Build a PyTorch model for this problem.
# Steps:
# • Prepare dataset
# • Define model
# • Choose loss function
# • Train model
# • Evaluate results
import torch
import torch.nn as nn
import torch.optim as optim

# Step 1: Prepare dataset (features: duration, budget, popularity)
inputs = torch.randn(100, 3)

# Simulated target (movie rating)
targets = (inputs[:,0]*2 + inputs[:,1]*3 + inputs[:,2]*1.5 + 5).unsqueeze(1)

# Step 2: Define model
model = nn.Sequential(
    nn.Linear(3, 6),
    nn.ReLU(),
    nn.Linear(6, 1)   # Output = single rating
)

# Step 3: Loss function (Regression)
criterion = nn.MSELoss()

# Step 4: Optimizer
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Step 5: Training
epochs = 100

for epoch in range(epochs):
    outputs = model(inputs)
    loss = criterion(outputs, targets)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# Step 6: Evaluation
with torch.no_grad():
    sample = torch.tensor([[0.5, 1.0, -0.2]])  # example movie features
    prediction = model(sample)
    print("\nPredicted Rating:", prediction.item())
```

```python
# 5. Game Winner Predictor

# Goal: Build a PyTorch model for this problem.
# Steps:
# • Prepare dataset

# • Define model
# • Choose loss function
# • Train model
# • Evaluate results
import torch
import torch.nn as nn
import torch.optim as optim

# Step 1: Prepare dataset (features: team1 strength, team2 strength, home advantage)
inputs = torch.randn(100, 3)

# Target: 1 if team1 wins, else 0
targets = ((inputs[:,0] + inputs[:,2]) > inputs[:,1]).float().unsqueeze(1)

# Step 2: Define model
model = nn.Sequential(
    nn.Linear(3, 5),
    nn.ReLU(),
    nn.Linear(5, 1),
    nn.Sigmoid()   # for binary classification
)

# Step 3: Loss function
criterion = nn.BCELoss()

# Step 4: Optimizer
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Step 5: Training
epochs = 100

for epoch in range(epochs):
    outputs = model(inputs)
    loss = criterion(outputs, targets)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# Step 6: Evaluation
with torch.no_grad():
    predictions = (model(inputs) > 0.5).float()
    accuracy = (predictions == targets).float().mean()
    print("\nAccuracy:", accuracy.item())
```

```python
# 10. Digit Classifier
# Goal: Build a PyTorch model for this problem.
# Steps:
# • Prepare dataset
# • Define model
# • Choose loss function
# • Train model
# • Evaluate results
import torch
import torch.nn as nn
import torch.optim as optim

# Step 1: Prepare dataset (28x28 image → flattened to 784 features)
inputs = torch.randn(100, 784)

# Targets: digits from 0 to 9
targets = torch.randint(0, 10, (100,))

# Step 2: Define model
model = nn.Sequential(
    nn.Linear(784, 128),
    nn.ReLU(),
    nn.Linear(128, 64),
    nn.ReLU(),
    nn.Linear(64, 10)   # 10 output classes
)

# Step 3: Loss function
criterion = nn.CrossEntropyLoss()

# Step 4: Optimizer
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Step 5: Training
epochs = 100

for epoch in range(epochs):
    outputs = model(inputs)
    loss = criterion(outputs, targets)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# Step 6: Evaluation
with torch.no_grad():
    predictions = torch.argmax(model(inputs), dim=1)
    accuracy = (predictions == targets).float().mean()
    print("\nAccuracy:", accuracy.item())
```
